```python
import pandas as pd
import numpy as np

import requests

#graph
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.cbook as cbook
```

```python
#Confirmed positive cases of COVID-19
url = "https://data.ontario.ca/dataset/f4112442-bdc8-45d2-be3c-12efae72fb27/resource/455fd63b-603d-4608-8216-7d8647f43350/download/conposcovidloc.csv"
filename = 'conposcovidloc.csv'
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
response = requests.get(url, headers=headers)
with open(filename, mode="wb") as file:
    file.write(response.content)
all_cases = pd.read_csv(filename)
```

```python
#COVID-19 testing percent positive by age group
url = "https://data.ontario.ca/dataset/ab5f4a2b-7219-4dc7-9e4d-aa4036c5bf36/resource/05214a0d-d8d9-4ea4-8d2a-f6e3833ba471/download/percent_positive_by_agegrp.csv"
filename = 'percent_positive_by_agegrp.csv'
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
response = requests.get(url, headers=headers)
with open(filename, mode="wb") as file:
    file.write(response.content)
pctpos_agegrp = pd.read_csv(filename)
```

```python
#COVID-19 cases in hospital and ICU
url = "https://data.ontario.ca/dataset/8f3a449b-bde5-4631-ada6-8bd94dbc7d15/resource/e760480e-1f95-4634-a923-98161cfb02fa/download/region_hospital_icu_covid_data.csv"
filename = 'region_hospital_icu_covid_data.csv'
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
response = requests.get(url, headers=headers)
with open(filename, mode="wb") as file:
    file.write(response.content)
icu_hosp = pd.read_csv(filename)
```

```python
#Availability of adult and pediatric ICU beds and occupancy for COVID-related critical illness
url = "https://data.ontario.ca/dataset/1b5ff63f-48a1-4db6-965f-ab6acbab9f29/resource/c7f2590f-362a-498f-a06c-da127ec41a33/download/icu_beds.csv"
filename = 'icu_beds.csv'
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
response = requests.get(url, headers=headers)
with open(filename, mode="wb") as file:
    file.write(response.content)
icu_bed = pd.read_csv(filename)
```

```python
#update date variables from char -> datetime64
#Accurate Episode Date
all_cases['Accurate_Episode_Date'] = pd.to_datetime(all_cases['Accurate_Episode_Date'])
all_cases['Accurate_Episode_Year'] = all_cases['Accurate_Episode_Date'].dt.year
all_cases['Accurate_Episode_Month'] = all_cases['Accurate_Episode_Date'].dt.month
#Case Reported Date
all_cases['Case_Reported_Date'] = pd.to_datetime(all_cases['Case_Reported_Date'])
all_cases['Case_Reported_Year'] = all_cases['Case_Reported_Date'].dt.year
all_cases['Case_Reported_Month'] = all_cases['Case_Reported_Date'].dt.month
#Test Reported Date
all_cases['Test_Reported_Date'] = pd.to_datetime(all_cases['Test_Reported_Date'])
#Specimen Date
all_cases['Specimen_Date'] = pd.to_datetime(all_cases['Specimen_Date'])

#pctpos_agegrp date
pctpos_agegrp['DATE'] = pd.to_datetime(pctpos_agegrp['DATE'])
#icu_hosp date
icu_hosp['date'] = pd.to_datetime(icu_hosp['date'])
#icu_bed date
icu_bed['date'] = pd.to_datetime(icu_bed['date'])
```

```python
#create recent data as of 2022-01-01, cut off at 2023-10-31
recent_cases = all_cases.loc[all_cases['Accurate_Episode_Date'] > '2022-10-01']
recent_cases = recent_cases.loc[all_cases['Accurate_Episode_Date'] <= '2023-10-31']

recent_pctpos_agegrp = pctpos_agegrp.loc[pctpos_agegrp['DATE'] > '2022-10-01']
recent_pctpos_agegrp = recent_pctpos_agegrp.loc[pctpos_agegrp['DATE'] <= '2023-10-31']

recent_icu_hosp = icu_hosp.loc[icu_hosp['date'] > '2022-10-01']
recent_icu_hosp = recent_icu_hosp.loc[icu_hosp['date'] <= '2023-10-31']

recent_icu_bed = icu_bed.loc[icu_bed['date'] > '2022-10-01']
recent_icu_bed = recent_icu_bed.loc[icu_bed['date'] <= '2023-10-31']
```

```python
recent_cases['Age_Group'].value_counts(normalize=True).plot(kind='bar')
```

```python
recent_cases_age_sex = recent_cases.groupby(['Age_Group'])['Client_Gender'].value_counts(normalize=True).to_frame(name='pct').reset_index()
recent_cases_age_sex
```

```python
#plot sex by age group

male = recent_cases_age_sex.loc[recent_cases_age_sex['Client_Gender'] == 'MALE'].sort_values('Age_Group')
female = recent_cases_age_sex.loc[recent_cases_age_sex['Client_Gender'] == 'FEMALE'].sort_values('Age_Group')

age_group = np.arange(len(female))

plt.figure(figsize=(10, 6))
plt.bar(age_group-0.15, male["pct"], label='Male', width=0.3)
plt.bar(age_group+0.15, female["pct"], label='Female', width=0.3)
plt.legend()
plt.xticks(ticks = age_group, labels = female['Age_Group'])
plt.title('Recent COVID-19 cases from October, 2022 by Age and Sex')
plt.xlabel('Age Group')
plt.ylabel('Proportion')
plt.show()

```

```python
agg = recent_cases.groupby(['Accurate_Episode_Date'])['Row_ID'].count().to_frame().reset_index()
agg['Accurate_Episode_Date'] = pd.to_datetime(agg['Accurate_Episode_Date'])
agg.tail()

fig = plt.figure()
ax = plt.axes()

x = agg['Accurate_Episode_Date']
y = agg['Row_ID']
ax.plot(x, y)

# Major ticks every quarter.
fmt_quarter = mdates.MonthLocator(interval=2)
ax.xaxis.set_major_locator(fmt_quarter)

# Minor ticks every 2 months.
fmt_month = mdates.MonthLocator()
ax.xaxis.set_minor_locator(fmt_month)

# Text in the x axis will be displayed in 'YYYY-mm' format.
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))

plt.title("Number of recent COVID-19 cases over time from October, 2022")
```
